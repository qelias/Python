```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
filepath = 'G:\\Mes Documents\\Documents\\Python\\Udemy-Python-for-Machine-Learning-and-Data-Science-Masterclass\\Notes\\05-Seaborn'
df = pd.read_csv(filepath+'\\dm_office_sales.csv')
df
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division',palette='Set2')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,0.5))
```
