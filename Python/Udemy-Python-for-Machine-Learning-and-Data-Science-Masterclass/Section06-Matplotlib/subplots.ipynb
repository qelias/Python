```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig,axes = plt.subplots(1,2)
```

```python
fig,axes = plt.subplots(2,2)
```

```python
a = np.linspace(0,10,11)
b=a**4
x = np.arange(0,10)
y=2*x
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(2,2)
axes[0,0].plot(x,y)
axes[1,1].plot(a,b)
axes[1,1].set_title('dzdzadaaz')
fig.suptitle('TITLE',fontsize=16)
plt.tight_layout()
fig.set_dpi(200)
fig.set_figwidth(5)
fig.set_figheight(5)
fig.savefig("mysubplot.pdf",bbox_inches='tight')
```
