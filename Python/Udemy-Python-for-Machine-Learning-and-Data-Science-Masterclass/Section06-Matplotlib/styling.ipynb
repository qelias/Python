```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,11)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_ylim(0,20)
ax.plot(x,x,label='x vs x',color='purple',lw=3,marker='s',ms=3,ls='--',markerfacecolor='red',markeredgewidth=10,markeredgecolor='orange')
ax.plot(x,np.exp(x),label='x vs x^2',color='#e6c017',lw=0.7)
lines = ax.plot(x,2*x,label='x vs 2x',color='blue',lw=2)

lines[0].set_dashes([5,5,2,2])
ax.legend(loc=(-0.5,0.5))
```
