```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,x[-1])
plt.ylim(0,y[-1])
plt.savefig('my_first_plot.pdf')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```
