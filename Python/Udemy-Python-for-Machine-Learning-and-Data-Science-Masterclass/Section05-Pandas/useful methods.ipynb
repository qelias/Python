```python
def yelp(price):
    if price < 10:
        return '$'
    elif 10<= price <= 30:
        return '$$'
    else:
        return '$$$'
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
df[ ['total_bill','tip'] ].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
```

```python
df['quality'] = df[ ['total_bill','tip'] ].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
```
